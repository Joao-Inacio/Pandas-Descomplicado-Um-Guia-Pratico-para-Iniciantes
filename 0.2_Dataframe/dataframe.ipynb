```python
colunas_categoricas = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']

for coluna in colunas_categoricas:
    #print(coluna)
    tem_interrogacao = df[coluna].str.contains("\?").any()
    if tem_interrogacao:
        print(f"Na coluna '{coluna}' existem valores '?'")
```

```python
df.isnull().sum()
```

```python
for coluna in colunas_categoricas:
    df[coluna] = df[coluna].replace(' ?', np.nan)
```

## Concatenação

```python
data = {
    'idvenda': [123, 374, 654, 345],
    'data': ['2024-01-19', '2024-01-20', '2024-02-21', '2024-02-25'],
    'idvendedor': [1, np.nan, 2, 3],
    'valor_total': [45.76, 102.34, 56.34, 34.21]
}
df1 = pd.DataFrame.from_dict(data)
df1
```

```python
data = {
    'idvenda': [546, 232, 789, 2345],
    'data': ['2024-01-22', '2024-01-22', '2024-01-23', '2024-01-24'],
    'idvendedor': [3, 2, 1, 2],
    'valor_total': [45.76, 102.34, 56.34, 34.21]
}
df2 = pd.DataFrame.from_dict(data)
df2
```

```python
df_complete = pd.concat([df1, df2], ignore_index=True)
df_complete
```

## Junção

```python
data = {
    'idvendedor': [1, 2, 3, 4],
    'nome': ['Julia', 'Paulo', 'Jéssica', 'Marcos']
}
df_vendedor = pd.DataFrame.from_dict(data)
df_vendedor
```

```python
df_complete.merge(
    df_vendedor,
    how='inner',
    left_on=['idvendedor'],
    right_on=['idvendedor']
)
```

```python
df_complete = df_complete.merge(
    df_vendedor,
    how='left',
    left_on=['idvendedor'],
    right_on=['idvendedor']
)
```

```python
df_complete.query("nome == 'Jéssica'")
```
