# **Dataframes**

```python
# Bibliotecas
import pandas as pd
import numpy as np
```

## Criação de dataframes

```python
data = [
    ['Alice', 25, 'São Paulo'],
    ['João', 30, 'Rio de Janeiro'],
    ['Bruno', 35, 'Curitiba'],
    ['Carlos', 40, 'Manaus']
]
```

```python
type(data)
```

```python
colunas = ['Nome', 'Idade', 'Cidade']
```

```python
df = pd.DataFrame(data=data, columns=colunas)
df
```

```python
data = {
    'Nome': ['Alice', 'João', 'Bruno', 'Carlos'],
    'Idade': [25, 30, 35, 40],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Manaus']
}
data
```

```python
type(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
data = np.array([
    ['Alice', 25, 'São Paulo'],
    ['João', 30, 'Rio de Janeiro'],
    ['Bruno', 35, 'Curitiba'],
    ['Carlos', 40, 'Manaus']
])
type(data)
```

```python
df = pd.DataFrame(data, columns=colunas)
df
```

```python
data = [
    ('Alice', 25, 'São Paulo'),
    ('João', 30, 'Rio de Janeiro'),
    ('Bruno', 35, 'Curitiba'),
    ('Carlos', 40, 'Manaus')
]
```

```python
type(data), type(data[0])
```

```python
df = pd.DataFrame(data, columns=colunas)
df
```

```python
data = {
    'Nome': pd.Series(['Alice', 'João', 'Bruno', 'Carlos']),
    'Idade': pd.Series([25, 30, 35, 40]),
    'Cidade': pd.Series(['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Manaus']
)}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
data = {
    'Nome': np.array(['Alice', 'João', 'Bruno', 'Carlos']),
    'Idade': np.array([25, 30, 35, 40]),
    'Cidade': np.array(['São Paulo', 'Rio de Janeiro', 'Curitiba', 'Manaus'])
}
data
```

```python
type(data)
```

```python
df = pd.DataFrame(data)
df
```
