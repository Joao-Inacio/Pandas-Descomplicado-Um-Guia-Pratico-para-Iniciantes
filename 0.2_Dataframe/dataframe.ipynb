### Valores faltantes

```python
df.isna().sum() #NaN
```

```python
df.loc[df['workclass'].str.contains("\?")]
```

```python
df['workclass'].str.contains('\?').sum()
```

```python
colunas_categoricas = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']

for coluna in colunas_categoricas:
    #print(coluna)
    tem_interrogacao = df[coluna].str.contains("\?").any()
    if tem_interrogacao:
        print(f"Na coluna '{coluna}' existem valores '?'")
```

```python
df.isnull().sum()
```

```python
for coluna in colunas_categoricas:
    df[coluna] = df[coluna].replace(' ?', np.nan)
```

```python
df.isna().sum()
```

```python
df.fillna("NOT INFORMED").iloc[0:30]
```

```python
df['workclass'].fillna(df['workclass'].mode().iloc[0], inplace=True)
df['occupation'].fillna(df['occupation'].mode().iloc[0], inplace=True)
df['native-country'].fillna(df['native-country'].mode().iloc[0], inplace=True)
```

```python
df.isna().sum()
```

## Renomear e reordenar

```python
df.columns = [coluna.upper() for coluna in df.columns]
df.columns
```

```python
df = df.reindex(labels=[
    'workclass', 'age', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain',
    'capital-loos', 'hour-per-week', 'native-country', 'income'
], axis=1)
df
```

## Criação de colunas

```python
df['hour-per-month'] = df['hour-per-week'] * 4
df
```

```python
df['high-education-level'] = df['education-num'] > 10
df
```

```python
dic_map = {" <=50K": "Low", " >50K": "High"}
df['category-income'] = df['income'].map(dic_map)
df.head()
```

```python
df['capital-gain-usd'] = df['capital-gain'].map(lambda x: f"USD {x}")
df.head()
```

```python
from datetime import datetime
datetime.now().year
```

```python
df = df.assign(
    birth=datetime.now().year - df["age"],
    hardwork=df["hour-per-week"] > 40
)
df
```

## Agrupamento com agregação

```python
df.groupby(['income', 'workclass'], observed=False).agg("mean", numeric_only=True)
```

```python
df.groupby([
    'income', 'workclass'
], observed=False).agg({
    "age": ["mean", "std"],
    "hour-per-week": ["min", "max"]
    })
```

```python
df.groupby([
    'income', 'workclass'
], observed=False).agg(
    age_mean=("age", "mean"),
    age_std=("age", "std"),
    hour_min=("hour-per-week", "min"),
    hour_max=("hour-per-week", "max")
)
```
