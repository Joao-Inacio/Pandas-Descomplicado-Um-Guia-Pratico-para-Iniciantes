- Criar uma função para atualizar o atributo User_Score, para deixá-lo na mesma escala do atributo Critic_Score (somente multiplicar por 10)
- Use a função apply para realizar a alteração

```python
def normaliza(score):
    return score * 10
```

```python
df['User_Score'] = df['User_Score'].apply(normaliza)
```

- Agrupamento de Platform por Global Sales

```python
agrupamento = df.groupby('Platform', observed=False)['Global_Sales'].mean()
agrupamento
```

- Agrupamento de Genre e Platform, mostrando a média dos outros atributos (função agg)

```python
agrupamento2 = df.groupby(['Platform', 'Genre'], observed=False).agg('mean', numeric_only=True)
agrupamento2
```

- A contagem de vendas por gênero

```python
agrupamento3 = df.groupby('Genre', observed=False)['Global_Sales'].size()
agrupamento3
```

- A média de Global Sales pelo gênero

```python
agrupamento3 = df.groupby('Genre', observed=False)['Global_Sales'].mean().reset_index()
agrupamento3
```

- Adicione uma nova coluna com a média do Score (função transform)

```python
df['avg_score'] = df.groupby('Genre', observed=False)['Global_Sales'].transform('mean')
jogos_destaque = df[df['Global_Sales'] > df['avg_score']]
jogos_destaque
```

- Crie uma tabela pivot para representar os dados conforme a tabela abaixo

```python
df.pivot_table(index='Genre', observed=False, columns='Platform', values='Global_Sales', aggfunc='mean')
```
